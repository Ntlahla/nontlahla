```python
from matplotlib import pyplot as plt
from matplotlib import style
 
style.use('ggplot') 
 
x = ['Simba','Lays','Doritos','Pringles','Cheetos'] 
y = [17,15,20,22,11] 
 
plt.bar(x, y, align='center') 
 
plt.title('Chips') 
plt.ylabel('price list of Chips in Rands') 
plt.xlabel('types of chips') 
 
plt.show() 
```
